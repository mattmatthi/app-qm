> **(b)** *Approximate the ground state of the one-dimensional harmonic oscillator using a trial wave function $\psi = \alpha^2 - x^2$ on $|x| \leq \alpha$, and $\psi = 0$ otherwise.*

$$
    \hat{H} = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2} + \frac 1 2 m\omega x^2
$$

$$
    \hat{H}\psi(x) = -\frac{\hbar^2}{2m}(-2) + \frac 1 2 m\omega x^2 (\alpha^2-x^2) = \frac{\hbar^2}{m} + \frac 1 2 m\omega x^2 (\alpha^2-x^2)
$$
\begin{align}
    \langle \psi | \hat{H} | \psi \rangle &= \int_{-\alpha}^{\alpha}(\alpha^2-x^2)\frac{\hbar^2}{m} + \frac 1 2 m\omega^2 x^2(\alpha^2-x^2) \ dx \\\ 
    &= \frac{\hbar^2}{m}\int_{-\alpha}^{\alpha}(\alpha^2-x^2)\ dx + \frac 1 2 m\omega^2\int_{-\alpha}^{\alpha} x^2(\alpha^2-x^2) \ dx \\\ 
    &= \frac{\hbar^2}{m}\left(2\alpha^3-\frac{2}{3}\alpha^3\right) + \frac 1 2 m\omega^2 \frac{16\alpha^7}{105} \\\ 
    &= \frac{4}{3}\frac{\hbar^2}{m}\alpha^3 + \frac{2}{15}m\omega^2\alpha^5
\end{align}

$$
    \langle \psi | \psi \rangle = \int_{-\alpha}^{\alpha}(\alpha^2-x^2)^2 \ dx = \frac{16\alpha^5}{15}
$$

$$
    \langle E_g \rangle \leqslant \frac{\langle \psi | \hat{H} | \psi\rangle}{\langle\psi | \psi\rangle} = \frac{5}{4}\frac{\hbar^2}{m}\frac{1}{\alpha^2} + \frac{1}{14}m\omega^2\alpha^2 \equiv \tilde{E_g}
$$
Minimizing,
$$
\frac{d\tilde{E_g}}{d\alpha} = -\frac 5 2 \frac{\hbar^2}{m}\frac{1}{\alpha^3} + \frac 1 7 m \omega^2 \alpha = 0 \Longrightarrow \alpha^4 = \frac{35}{2}\frac{\hbar^2}{m}
$$
We then have an upper bound to the ground state,
$$
    \langle E_g \rangle \leqslant \left(\frac 5 4 \sqrt{\frac{2}{35}} + \frac{1}{14}\sqrt{\frac{35}{2}}\right)\hbar\omega \approx 0.598\hbar\omega
$$

> **(a)** *Approximate the ground-state wave function by $\langle \mathbf{r}|\psi\rangle$.*

In atomic units, 
$$
    \hat{H} = \left(-\frac{\nabla^2}{2} - \frac 1 r\right)
$$

\begin{align}
 \hat{H} \psi(\mathbf{r}) &= -\frac{1}{r^2}\frac{d}{dr}\left(r^2\frac{d\psi}{dr}\right) - \frac \psi r
 = -\frac{1}{r^2}\frac{d}{dr}\left(-2\alpha r^3\psi\right) - \frac{\psi}{r}
 = \frac{2\alpha}{r^2}\left(3r^2\psi - 2\alpha r^4 \psi\right) - \frac{\psi}{r}
 = \left( 6\alpha - 4\alpha r^2 - \frac{1}{r}\right) e^{-\alpha r^2}
\end{align}

$$
 \langle E_g \rangle \leqslant \tilde{E_g} = \frac{\langle \psi | \hat{H} | \psi\rangle}{\langle\psi|\psi\rangle} = \frac{\int \left(6\alpha - 4\alpha r^2 - \frac{1}{r}\right)e^{-2\alpha r^2} d^3 r}{\int e^{-2\alpha r^2}d^3 r} = 6\alpha - 4\alpha\frac{\int r^2 e^{-2\alpha r^2}d^3 r}{\int e^{-2\alpha r^2}d^3 r} - \frac{\int re^{-2\alpha r^2}d^3 r}{\int e^{-2\alpha r^3}d^3 r} = 6\alpha - 4\alpha I_1 - \frac{I_2^{(1)}}{I_2^{(2)}}
$$

Let $\mathcal{I}(a) = \int e^{-a r^2} d^3 r$. Then
$$I_1 = -\frac{d\ln \mathcal{I}(2\alpha)}{d(2\alpha)} = -\frac{1}{2}\frac{d\ln \left(\frac{1}{2\alpha}\right)^{3/2}\mathcal{I}(1)}{d\alpha} = \frac{3}{4\alpha}$$

The last integrals we calculate explicitly:

\begin{align}
I_2^{(1)} &= \int \frac{1}{r}e^{-2\alpha r^2} d^3 r = 4\pi\int_0^\infty r e^{-2\alpha r^2} dr = -\frac{\pi}{\alpha}\int_0^\infty \frac{d}{dr}(e^{-2\alpha r^2}) dr = \frac{\pi}{\alpha}\\\ 
I_2^{(2)} &= \int e^{-2\alpha r^2} d^3 r = 4\pi\int_0^\infty r^2 e^{-2\alpha r^2} dr = 4\pi \frac{d}{d(-2\alpha)}\int_0^\infty e^{-2\alpha r^2} dr = 4\pi \frac{d}{d(-2\alpha)} \sqrt{\frac{\pi}{2\alpha}} = \frac{1}{2\sqrt{2}}\left(\frac{\pi}{\alpha}\right)^{3/2}
\end{align}

Putting it all together,

\begin{align}
 \tilde{E_g} = 6\alpha -\alpha I_1 - \frac{I_2^{(1)}}{I_2^{(2)}} = 6\alpha - 4\alpha \frac{3}{4\alpha} - \frac{\pi/\alpha}{(\pi/\alpha)^{3/2}/(2\sqrt{2})}
 = 3(2\alpha - 1) - 2\sqrt{\frac{2\alpha}{\pi}}
\end{align}

Minimizing,

$$
    \frac{d\tilde{E_g}}{d\alpha} = 6-\left(\frac{2\alpha}{\pi}\right)^{-1/2}\frac{2}{\pi} = 0 \Longrightarrow \alpha = \frac{1}{18\pi}
$$

Hence the upper bound is
$$\langle E_g\rangle \leqslant \tilde{E_g}\left(\frac{1}{18\pi}\right) = -\frac{1}{3\pi} - 3 \approx -3.106 \ Ry$$

which is pretty far from the exact ground state energy $-1 \ Ry$.

To find the matrix representation of the Hamiltonian (for the subspace), we calculate the components directly:

\begin{align*}
  \hat{H} \ket{\psi_n} &= \frac 1 {2\pi} \left(-E_c\frac{d^2}{d\delta^2} - E_J\cos\delta\right)e^{in\delta}
                    = \frac 1 {2\pi}\left(E_C n^2 - E_J \cos\delta\right)\ket{\psi_n} \\\ \\\ 
  \bra{\psi_m}\hat{H}\ket{\psi_n}
                    &= \frac{1}{2\pi}\int_0^{2\pi}\left(E_C n^2 - E_J\cos\delta\right)e^{i(n-m)\delta d\delta} \\\ 
                    &= \frac 1 {2\pi} \left(E_C n^2 \int_0^{2\pi}e^{i(n-m)\delta}d\delta - E_J\int_0^{2\pi}e^{i(n-m)\delta}\cos(\delta)d\delta\right)
                    \equiv H_{mn}
\end{align*}

If $m = n$, $H_{nn} = E_C n^2$. If $m \neq n$,

\begin{align*}
  H_{mn} &= -\frac{1}{2\pi}\int_0^{2\pi}e^{i(m-n)\delta}\cos\delta d\delta
         = -\frac{1}{2\pi}\int_0^{2\pi}e^{i(m-n)\delta}\frac{e^{i\delta}+e^{-i\delta}}{2}d\delta
         = -\frac{1}{4\pi}\int_0^{2\pi} \left(e^{i(m-n+1)\delta} + e^{i(m-n-1)\delta} \right) d\delta
\end{align*}

If $m-n+1 = 0$ or $m-n-1 = 0$, that is, adjacent to the diagonal, $H_{mn} = -\frac{E_J}{2}$. Otherwise $H_{mn} = 0$. Putting this in matrix form,

$$ H = \begin{pmatrix} 
    E_C & -\frac 1 2 E_J & 0 \\\ 
    -\frac 1 2 E_J & 0 & -\frac 1 2 E_J \\\ 
    0 & -\frac 1 2 E_J & E_C
  \end{pmatrix}
$$

```python
import numpy as np


# The Josephson junction Hamiltonian
def JJ_hamiltonian(n,Ec,Ej):
      if n%2 == 0:
          raise ValueError('The Hamiltonian must be of odd dimension')
      else:
          H = np.zeros((n,n))
          v = symmetric_array(n)
          for i in range(n):
              H[i][i] = Ec*v[i]**2
              if i > 0:
                  H[i][i-1] = -Ej/2
              if i < n-1:
                  H[i][i+1] = -Ej/2
          return H

# Returns an array of the form -n,...,0,...n
def symmetric_array(n):
    v_pos = [i for i in range(int(np.floor(n/2))+1)]
    v_neg = [-x for x in reversed(v_pos)]
    return np.append(v_neg[:-1],[v_pos])

# Returns k energy levels in ascending order
def energies(H,k):
    eigenvals = np.sort(np.linalg.eigvalsh(H))
    return eigenvals[0:k]

print("The upper bound for the ground state energy is", energies(JJ_hamiltonian(99,0.2,0.6),1)[0])
```

> **(c)** *Consider $E_C = 10E_J$ . Restrict your trial functions to the subspace $n = −N,\ldots,0,\ldots,N$. Plot the ground and ﬁrst-excited state energies as a function of $n_g$ in the range  $n_g \in [−1,1]$. Do this for $N = 1$, $N = 2$ and $N = 3$.*

```python
# The hamiltonian for the Cooper-pair box
def CB_hamiltonian(n,Ec,Ej,ng):
    H = np.zeros((n,n))
    v = symmetric_array(n)
    for i in range(n):
      H[i][i] = 4*Ec*offset(v[i],ng)
      if i > 0:
          H[i][i-1] = -Ej/2
      if i < n-1:
          H[i][i+1] = -Ej/2
    return H

def offset(p,ng):
    return (p-ng)**2
```
